```python
import numpy as np
import matplotlib.pyplot as plt

# Définir les constantes
C = 7
A = 30

# Créer une gamme de valeurs pour D
D = np.linspace(100, 10, 400)  # Évite D=0 pour éviter la division par zéro

# Calculer la fonction (A - C) * 100 * np.log(C) / D
function_values = (A - C) * 100 * np.log(C) / D

# Tracer la fonction
plt.plot(D, function_values, label=r'$(A - C) \cdot \frac{100 \cdot \log(C)}{D}$')
plt.xlabel('D')
plt.ylabel('Function Value')
plt.title('Plot of $(A - C) \cdot \\frac{100 \cdot \\log(C)}{D}$')
plt.grid(True)
plt.legend()
plt.show()

```

## Creating a function

```python
def choisir_mon_cpl (age, age_couple, distance_km) :  
    if ((age - age_couple) * np.log(age_couple^100) / distance_km)<1 :
        print(f"Il vaut mieux choisir le couple")
    else :
        print(f"Il vaut mieux choisir la ville")
```
